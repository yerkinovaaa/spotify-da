#### In which year were the most hits released?

```python
spotify_data['released_year'].value_counts().idxmax()
```

```python
filter = spotify_data[spotify_data['released_year'] >= 2010]
```

##### released hits starting from 2010

```python
hits_per_year = filter['released_year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
hits_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Hits Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Hits')
plt.xticks(rotation=45)
plt.show()
```

#### Grouping by decades

```python
spotify_data['by_decade'] = (spotify_data['released_year'] // 10) * 10
hits_per_decade = spotify_data['by_decade'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
hits_per_decade.plot(kind='bar', color='skyblue')
plt.title('Number of Hits Released Per Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Hits')
plt.xticks(rotation=45)
plt.show()
```

#### Top Years

```python
top_years = spotify_data['released_year'].value_counts().nlargest(10).sort_index()  # Top 10 years
plt.figure(figsize=(10, 6))
top_years.plot(kind='pie')
plt.title('Top 10 Years with Most Hits Released')
plt.xlabel('Year')
plt.ylabel('Number of Hits')
plt.xticks(rotation=45)
plt.show()
```

#### Top 3 Artists by Number of Songs in Chart

```python
top_3_artists = spotify_data['artist(s)_name'].value_counts().head(3)

top_3_artists.plot(kind='barh', color='gold')
plt.title('Top 3 Artists by Number of Songs in Chart')
plt.xlabel('Number of Songs')
plt.ylabel('Artist')
plt.show()
```
